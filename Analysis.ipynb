```python
for j, entry in enumerate(neo_data):
    t = [i for i in range(1, len(entry)+1)]
    print("-"*50)
    print(tests[j])
    print("-"*50)
    sns.relplot(
        data=entry,
        x=t,
        y="cpu",
        kind="line"
    )
    sns.relplot(
        data=entry,
        x=t,
        y="mem",
        kind="line"
    )
    sns.relplot(
        data=entry,
        x=t,
        y="pid_count",
        kind="line"
    )
    plt.show()
```
